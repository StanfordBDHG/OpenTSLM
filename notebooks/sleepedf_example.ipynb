# Sleep-EDF Dataset Example

This notebook demonstrates how to load the Sleep-EDF dataset using `SleepEDFDataset` and plot an example epoch.

```python
import os
import sys
sys.path.append(os.path.dirname(os.getcwd()))
# Adjust path to the loader module
from time_series_datasets.sleep_edf.sleepedf_loader import SleepEDFDataset, get_sleepedf_data
```

```python
import matplotlib.pyplot as plt
```

```python
data_path = get_sleepedf_data()
dataset = SleepEDFDataset(data_dir=data_path)
# Get the first recording's epochs and labels
data, labels = dataset[0]
# Select and plot the first epoch
epoch = data[0, 0, :]
label = labels[0]
plt.figure(figsize=(12, 4))
plt.plot(epoch)
plt.title(f'Epoch 0 - Label {label}')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()
```
